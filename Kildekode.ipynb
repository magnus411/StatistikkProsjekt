```python
#Importerer nødvendige biblioteker
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import statsmodels.formula.api as smf
import statsmodels.api as sm


```

```python
# Leser inn datasettet
df = pd.read_csv("lego.population.csv", sep=",", encoding="latin1")



# Fjerner kolonner vi ikke trenger
df2 = df[
    ["Set_Name", "Theme", "Gender", "Ages", "Price", "Pages", "Pieces",  "Unique_Pieces"]
]

# fjerner observasjoner med manglende datapunkter
df2 = df2.dropna()


# gjør themes om til string og fjern alle tegn vi ikke vil ha med
df2["Theme"] = df2["Theme"].astype(str)
df2["Theme"] = df2["Theme"].str.replace(r"[^a-zA-Z0-9\s-]", "", regex=True)

# fjerner dollartegn og trademark-tegn fra datasettet
df2["Price"] = df2["Price"].str.replace("\$", "", regex=True)

# og gjør så prisen om til float
df2["Price"] = df2["Price"].astype(float)



df2
```

```python
# Enkel lineær regresjon: Pris som en funksjon av kjønn
formel = 'Price ~ Gender'

modell = smf.ols(formel, data = df2)
resultat = modell.fit()
resultat.summary()
```

```python
#Separerer unike kjønn
gender_groups = df2["Gender"].unique()



# Utfør lineær regresjon og lag scatterplott med regresjonslinjer for hver kjønnsgruppe
for gender_group in gender_groups:

    # Filtrer datasettet for den aktuelle kjønnsgruppen
    subset_df = df2[df2['Gender'] == gender_group]
    
    # Opprett en lineær regresjonsmodell: Pris som en funksjon av antall brikker
    modell = smf.ols('Price ~ Pieces', data=subset_df)
    result = modell.fit()

    # Hent stigningstall og skjæringspunkt for regresjonslinjen
    slope = result.params['Pieces']
    intercept = result.params['Intercept']

    # Generer X og Y-verdier for regresjonslinjen
    regression_x = np.array(subset_df['Pieces'])
    regression_y = slope * regression_x + intercept

    # Lag et scatterplott og tegn regresjonslinjen
    plt.scatter(subset_df['Pieces'], subset_df['Price'], color=plt.cm.tab10(0))
    plt.plot(regression_x, regression_y, color="red", label=gender_group)

    # Sett aksenavn og tittel for plottet
    plt.xlabel('Antall brikker')
    plt.ylabel('Pris [$]')
    plt.title('Kryssplott med regresjonslinje (enkel LR)')
    plt.legend()
    plt.grid()

    # Vis plottet
    plt.show()

    #Printer resultatet
    print(result.summary())
   
```

```python
# Enkel lineær regresjon: Pris som en funksjon av Pieces
formel = 'Price ~ Pieces'

modell = smf.ols(formel, data = df2)
resultat = modell.fit()
resultat.summary()

```

```python

# plott av dataene og regresjonslinjene
slope = resultat.params['Pieces']
intercept = resultat.params['Intercept']


regression_x = np.array(df2['Pieces'])
regression_y = slope * regression_x + intercept

# Plot scatter plot and regression line
plt.scatter(df2['Pieces'], df2['Price'], label='Data Points')
plt.plot(regression_x, regression_y, color='red', label='Regression Line')
plt.xlabel('Antall brikker')
plt.ylabel('Pris [$]')
plt.title('Kryssplott med regresjonslinje (enkel LR)')
plt.legend()
plt.grid()

# Vis plottet
plt.show()
```

```python
# Enkel lineær regresjon: Pris som en funksjon av Unique_Pieces
formel = 'Price ~ Unique_Pieces'

modell = smf.ols(formel, data = df2)
resultat = modell.fit()
resultat.summary()
```

```python
# Plot scatter plot and regression line
slope = resultat.params['Unique_Pieces']
intercept = resultat.params['Intercept']

regression_x = np.array(df2['Unique_Pieces'])
regression_y = slope * regression_x + intercept

# Plot scatter plot and regression line
plt.scatter(df2['Unique_Pieces'], df2['Price'], label='Data Points')
plt.plot(regression_x, regression_y, color='red', label='Regression Line')
plt.xlabel('Antall unike brikker')
plt.ylabel('Pris [$]')
plt.title('Kryssplott med regresjonslinje (enkel LR)')
plt.legend()
plt.grid()
plt.show()
```

```python
# multippel lineær regresjon 
# 'Pieces' og 'Gender', samt deres interaksjon ('Pieces * Gender'), er de uavhengige variablene
modell3_mlr = smf.ols('Price ~ Pieces * Gender', data=df2)
results_mlr = modell3_mlr.fit()
summary_mlr = results_mlr.summary()
summary_mlr

```

```python
gender_groups = df2["Gender"].unique()

resultater = []

for i, gender in enumerate(gender_groups):
    modell3 = smf.ols('Price ~ Pieces', data=df2[df2['Gender'] == gender])
    resultater.append(modell3.fit())

# Plot the data and regression lines for different genders
for i, gender in enumerate(gender_groups):
    slope = resultater[i].params['Pieces']
    intercept = resultater[i].params['Intercept']

    regression_x = np.array(df2[df2['Gender'] == gender]['Pieces'])
    regression_y = slope * regression_x + intercept

    # Plot scatter plot and regression line with semi-transparent dots
    plt.scatter(
        df2[df2['Gender'] == gender]['Pieces'],
        df2[df2['Gender'] == gender]['Price'],
        color=plt.cm.tab10(i),
        label=gender,
        alpha=0.2  # Set the transparency (0.0 to 1.0)
    )
    plt.plot(regression_x, regression_y, color=plt.cm.tab10(i))

plt.xlabel('Antall brikker')
plt.ylabel('Pris')
plt.title('Kryssplott med regresjonslinjer for ulike kjønn')

# Set custom y-axis limits
plt.ylim(0, 500)  # Adjust the limits as needed
plt.xlim(0, 6000)  # Adjust the limits as needed

plt.legend()
plt.grid()
plt.show()
```
